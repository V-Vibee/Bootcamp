**Import Dependencies**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('https://github.com/codebasics/py/raw/master/ML/FeatureEngineering/3_outlier_IQR/heights.csv')
df.head()
```

```python
df.describe()
```

**Detect Outlier using IQR**

```python
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python

lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

**Outliers values**

```python
df[(df.height<lower_limit)|(df.height>upper_limit)]
```

**Remove Outliers**

```python
df_no_outlier = df[(df.height>lower_limit)&(df.height<upper_limit)]
df_no_outlier
```
