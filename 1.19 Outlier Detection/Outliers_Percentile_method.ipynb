**Import Dependencies**

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://github.com/codebasics/py/raw/master/ML/FeatureEngineering/1_outliers/heights.csv')
df.head()
```

```python
df['height'].quantile()
```

```python
max_threshold= df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height']> max_threshold]
```

```python
min_threshold= df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']< min_threshold]
```

```python
df[(df['height']< max_threshold) & (df['height']> min_threshold)]
```

#**2. Another Example**

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://github.com/codebasics/py/raw/master/ML/FeatureEngineering/1_outliers/bhp.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
min_threshold, max_threshold= df.price_per_sqft.quantile([0.001, 0.999])
min_threshold, max_threshold
```

```python
df[df['price_per_sqft']< min_threshold]
```

```python
df[df.price_per_sqft > max_threshold]
```

**Remove Outliers**

```python
df2 = df[(df.price_per_sqft<max_threshold) & (df.price_per_sqft>min_threshold)]
df2.shape
```
