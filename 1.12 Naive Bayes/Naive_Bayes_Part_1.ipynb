**Import Dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('https://github.com/codebasics/py/raw/801ee0ee4d342fd22b89915dc0c4864250a0ec10/ML/14_naive_bayes/titanic.csv#L4')
df.head()
```

**Drop Irrelevant Columns**

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis=1, inplace=True)
df.head()
```

```python
target= df.Survived
inputs= df.drop('Survived', axis=1)
```

**Create Dummy columns**

```python
dummies= pd.get_dummies(inputs.Sex)
dummies.head()
```

**Concat dummy columns with input**

```python
inputs= pd.concat([inputs, dummies], axis=1)
inputs.head()
```

```python
# drop original column

inputs.drop('Sex', axis=1, inplace=True)
inputs.head()
```

```python
inputs.isna().any()
```

```python
inputs.Age= inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(inputs,target, test_size=0.2, random_state=2529)
```

```python
print(len(inputs))
print(len(X_train))
print(len(X_test))
```

**Naive Bayes Model**

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
