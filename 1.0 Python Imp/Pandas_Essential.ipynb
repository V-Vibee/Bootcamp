#**Filtering & Ordering**

```python
df[df['Rank'] < 10]
```

```python
specific_countries= ['Bangladesh', 'Brazil']
df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('United')]
```

```python
# loc: Location
# iloc: Integer Location
```

```python
df2= df.set_index('Country')
df2
```

```python
df2.filter(items= ['Continent', 'CCA3'])
```

```python
df2.filter(like= 'United', axis=0)
```

```python
df2.loc['United States']
```

```python
df2.iloc[3]
```

```python
df[df['Rank'] > 10].sort_values(by= ['Country','Continent'], ascending=[False,True])
```

#**Join, Merge & Concatenate**

```python
import pandas as pd
```

```python
df1= pd.read_csv('https://github.com/AlexTheAnalyst/PandasYouTubeSeries/raw/main/LOTR.csv')
```

```python
df1.head()
```

```python
df2= pd.read_csv('https://github.com/AlexTheAnalyst/PandasYouTubeSeries/raw/main/LOTR%202.csv')
df2.head()
```

```python
df1.merge(df2, how='inner')
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix= '_Left',rsuffix='_Right')
```

```python
df4= df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix= '_Left',rsuffix='_Right')
df4
```

**Concatenate**

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join='inner', axis=1)
```
