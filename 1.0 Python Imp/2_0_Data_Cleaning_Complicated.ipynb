```python
import pandas as pd
```

```python
df= pd.read_excel('https://github.com/AlexTheAnalyst/PandasYouTubeSeries/raw/main/Customer%20Call%20List.xlsx', na_values="N/a")
```

```python
df.head()
```

**Step1: Drop the duplicates/ Column**

```python
df= df.drop_duplicates()
```

```python
df= df.drop(columns='Not_Useful_Column')
```

**Step 2: Standardize the Last Name**

```python
df.head()
```

```python
df['Last_Name']= df['Last_Name'].str.strip("123./_")
df['Last_Name']
```

**Step 3: Standardize the Phone Number**

```python
df['Phone_Number'].unique()
```

```python

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x) if  isinstance(x, int)else x)
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('[^\d]', '', regex=True)
```

```python
df
```

**Now give the specific Format**

```python
df['Phone_Number'] =df['Phone_Number'].apply(lambda x: f'{x[0:3]}-{x[3:6]}-{x[6:10]}' if isinstance(x,str)else x)
```

```python
df.dropna(subset='Phone_Number', inplace=True)
```

```python
df.head()
```

**Handling Address Column**

```python
df[['street', 'state', 'zip_code']] = df['Address'].str.split(',', n=2, expand=True)
```

```python
df.drop('Address', axis=1, inplace=True)
```

```python
df
```

**Handling Paying Customer Column**

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('N')
```

```python
df
```

```python
df.fillna('', inplace=True)
```

```python
df
```

```python
df.drop(df[df['Do_Not_Contact'] == 'Y'].index, inplace=True)
```

```python
df
```

**Finally we need to reset index**

```python
df.reset_index(drop=True, inplace=True)
df
```
