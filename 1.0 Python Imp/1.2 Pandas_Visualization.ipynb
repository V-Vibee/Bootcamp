**Data Visualizations**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('https://github.com/AlexTheAnalyst/PandasYouTubeSeries/raw/main/Ice%20Cream%20Ratings.csv')
df.set_index('Date')
df.head()
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8')
```

```python
df.plot(kind='line', subplots= True );
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot.barh( stacked=True)
```

```python
df.plot.scatter(x='Texture Rating', y= 'Overall Rating', s=100, c='orange');
```

```python
df.plot.hist(bins=20);
```

```python
df.boxplot();
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y= 'Flavor Rating')
```
