#**Dummy Variable method**

**Import Dependencies**

```python
import pandas as pd
import numpy as np
```

**Import Dataset**

```python
df= pd.read_csv('https://github.com/codebasics/py/raw/801ee0ee4d342fd22b89915dc0c4864250a0ec10/ML/5_one_hot_encoding/homeprices.csv#L4')
df.head()
```

```python
dummies= pd.get_dummies(df.town)
dummies
```

```python
merged= pd.concat([df, dummies], axis=1)
merged
```

```python
# Drop original column now as well as one column of dummy, so avoid dummy trap

final= merged.drop(['town','west windsor'], axis=1)
final
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

**Define X Features & y Target**

```python
X= final.drop('price', axis=1)
y= final.price
```

```python
lr.fit(X,y)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
lr.predict([[2800,0,1]])
```

```python
lr.predict([[3400,0,0]])
```

```python
lr.score(X,y)
```

#**One Hot Encoding method**

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
dfle =df
dfle.town= le.fit_transform(dfle.town)
dfle
```

```python
# We use .values to hv 2 dimensional array & not dataframe

X = dfle[['town','area']].values
X
```

```python
y= dfle.price.values
y
```

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
# To drop tyhe index and take from 1st column
X= X[:,1:]
X
```

```python
lr.fit(X,y)
```

```python
lr.predict([[0,3400]]) # 3400 sqr ft home in west windsor
```

```python
lr.predict([[0,2800]]) # 2800 sqr ft home in robbinsville
```
