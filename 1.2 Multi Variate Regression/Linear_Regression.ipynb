```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df= pd.read_csv('https://github.com/codebasics/py/raw/801ee0ee4d342fd22b89915dc0c4864250a0ec10/ML/2_linear_reg_multivariate/homeprices.csv#L4')
df.head()
```

```python
df.isnull().sum()
```

```python
import math
median_bedrooms= math.floor(df.bedrooms.median())
median_bedrooms
```

```python
# Filling Missing Values
df.bedrooms =df.bedrooms.fillna(median_bedrooms)
df
```

```python
model= linear_model.LinearRegression()
model.fit(df[['area', 'bedrooms', 'age']], df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df['experience']= df['experience'].fillna('zero')
df
```

**Converting Word to Number**

```python
# Changing word to number
from word2number import w2n
df['experience']= df['experience'].apply(w2n.word_to_num)
```

```python
df
```

```python
# Handling test Score
import math
mean_score= math.floor(df['test_score(out of 10)'].mean())
mean_score
```

```python
df['test_score(out of 10)']= df['test_score(out of 10)'].fillna(mean_score)
df
```

```python
model= linear_model.LinearRegression()
model.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']], df['salary($)'])
```

```python
model.predict([[2,9,6]])
```
